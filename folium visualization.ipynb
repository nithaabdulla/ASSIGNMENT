```python

# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))

# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

def assign_marker_color(class_code):
    if class_code == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color']=spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

coastline_lat = 28.5635
coastline_lon = -80.56806

distance_coastline = calculate_distance(
    launch_sites_df['Lat'][0], launch_sites_df['Long'][0], coastline_lat, coastline_lon)
distance_coastline
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

coastline_coordinate = [coastline_lat, coastline_lon]

distance_marker = folium.Marker(coastline_coordinate, 
                                icon=DivIcon(icon_size=(20,20), 
                                             icon_anchor=(0,0), 
                                             html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline)))

site_map.add_child(distance_marker)
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)

coordinates = [(launch_sites_df['Lat'][0], launch_sites_df['Long'][0]), (coastline_lat, coastline_lon)]
lines=folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

# calculate the distances
railway_lat, railway_long = 28.57403, -80.58566
highway_lat, highway_long = 28.5662, -80.57259
city_lat, city_long = 28.60622, -80.8306

distance_railway = calculate_distance(launch_sites_df['Lat'][0], launch_sites_df['Long'][0], railway_lat, railway_long)
distance_highway= calculate_distance(launch_sites_df['Lat'][0], launch_sites_df['Long'][0], highway_lat, highway_long)
distance_city= calculate_distance(launch_sites_df['Lat'][0], launch_sites_df['Long'][0], city_lat, city_long)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
distance_marker_railway = folium.Marker(location=[railway_lat, railway_long], 
                                        icon=DivIcon(icon_size=(20,20), 
                                                     icon_anchor=(0,0), 
                                                     html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_railway)))
site_map.add_child(distance_marker_railway)

distance_marker_highway = folium.Marker(location=[highway_lat, highway_long], 
                                        icon=DivIcon(icon_size=(20,20), 
                                                     icon_anchor=(0,0), 
                                                     html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_highway)))
site_map.add_child(distance_marker_highway)

distance_marker_city = folium.Marker(location=[city_lat, city_long], 
                                        icon=DivIcon(icon_size=(20,20), 
                                                     icon_anchor=(0,0), 
                                                     html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_city)))
site_map.add_child(distance_marker_city)
```

```python
# Draw a line between the marker to the launch site
line_railway=folium.PolyLine(
    locations=[(launch_sites_df['Lat'][0], launch_sites_df['Long'][0]), (railway_lat, railway_long)], weight=1)
site_map.add_child(line_railway)

line_highway=folium.PolyLine(
    locations=[(launch_sites_df['Lat'][0], launch_sites_df['Long'][0]), (highway_lat, highway_long)], weight=1)
site_map.add_child(line_highway)

line_city=folium.PolyLine(
    locations=[(launch_sites_df['Lat'][0], launch_sites_df['Long'][0]), (city_lat, city_long)], weight=1)
site_map.add_child(line_city)
```
